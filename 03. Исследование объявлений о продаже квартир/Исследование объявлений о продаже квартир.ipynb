# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

###### Описание данных
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)
- `cityCenters_nearest` — расстояние до центра города (м)
- `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
- `first_day_exposition` — дата публикации
- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении

### Замена пропусков

Прежде чем менять типы данных в столбцах, нужно разобраться с пропусками в них, иначе программа будет выдавать ошибку.

```python
#выведем общую информацию о пропусках
df.isna().sum()
```

Пропуски есть в большей половине столбцов. Заменим их в тех столбцах, которые нам потребуются для анализа и повлияют на результаты исследования.

#### Высота потолков

```python
#посмотрим какие значения есть в этом столбце
df['ceiling_height'].unique()
```

Как видим, в данном столбце помимо пропусков есть и аномальные значения, включающие в себя и 1 метр и 100 метров.
Согласно строительным нормам высота потолков не может быть ниже 2,5 метров и получается, что значения ниже 2,5 метров являются некорректными, как и значения более 6 метров (двухуровневый пентхаус). Эти аномальные значения лучше заменить медианами.

```python
#посмотрим количество таких аномальных значений
len(df.query('ceiling_height < 2.5 or ceiling_height > 6'))
```

```python
#проверим медиану
display(df['ceiling_height'].median())
#и заменим пропуски медианными значениями высоты потолков
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

#### Количество этажей в доме

Пропущено небольшое количество значений в этом столбце, всего 86.

```python
#посмотрим среднее, минимальное и максимальное значения
display(df['floors_total'].median())
df['floors_total'].describe()
```

```python
#выведем первые 10 строк, чтобы посмотреть на каких этажах квартиры, в которых пропущены значения этажности дома
display(df[df['floors_total'].isna()].head(10))
```

Как мы видим, у нас есть этажи, которые выше среднего/медианного, значит заменить все пропуски на средние/медианные значения не получится.

```python
#заполним пропуски самыми популярными этажностями домов исходя из этажа нахождения квартиры
df.loc[(df['floors_total'].isna()) & (df['floor'] <=5), 'floors_total'] = 5
df.loc[(df['floors_total'].isna()) & (df['floor'] > 5)&(df['floor'] <=9), 'floors_total'] = 9
df.loc[(df['floors_total'].isna()) & (df['floor'] > 9)&(df['floor'] <=12),'floors_total'] = 12
df.loc[(df['floors_total'].isna()) & (df['floor'] > 12)&(df['floor'] <=16),'floors_total'] = 16
df.loc[(df['floors_total'].isna()) & (df['floor'] > 16)&(df['floor'] <=25),'floors_total'] = 25
#проверим количество пропусков в этом столбце
df['floors_total'].isna().sum()
```

Как видим, все пропуски заменены.

#### Жилая площадь и площадь кухни

Пропуски в данных столбцах заполним медианами исходя из общей площади квартиры и количества комнат в ней.

```python
#сначала для жилой площади
#здесь придется ввести временную переменную, которую удалим после замены пропусков
df['temp_living'] = df['living_area']/df['total_area']
df['living_area'] = df['living_area'].fillna(df.groupby('rooms')['temp_living'].transform('median'))
#проверим количество пропусков
df['living_area'].isna().sum()
```

```python
#теперь то же самое проделаем с площадью кухни
df['temp_kitchen'] = df['kitchen_area']/df['total_area']
df['kitchen_area'] = df['kitchen_area'].fillna(df.groupby('rooms')['temp_kitchen'].transform('median'))
#проверим количество пропусков
df['kitchen_area'].isna().sum()
```

У нас осталось 197 пропусков. Посмотрим, что это за строки и почему они остались.

```python
df.loc[df['kitchen_area'].isna()]
```

Исходя из данных остались строки с пропусками в тех квартирах, где маленькая площадь и количество комнат равняется нулю. Видимо это студии и небольшая часть квартир, в которых продавцы не указали, что это студии и в которых нет выделенной кухни и почти вся площадь является жилой. Заменим оставшиеся значения нулями.

```python
df['studio'].value_counts()
```

```python
df['kitchen_area'] = df['kitchen_area'].fillna(0)
#снова проверим количество пропусков
df['kitchen_area'].isna().sum()
```

```python
#и удалим ненужные столбцы, которые вводились для замены данных
df.drop(columns = ['temp_kitchen', 'temp_living'],axis = 1,inplace=True)
```

#### Апартаменты

В этом столбце содержится 20924 пропусков. Проверим какие значения есть в этом столбце.

```python
df['square_meter'] = df['last_price'] /df['total_area']
```

### День недели, месяц и год публикации объявления

Добавим отдельные столбцы с днем, месяцем и годом публикации объявления. Для этого воспользуемся методом datetime.

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

```python
#переименуем столбец с днями недели для удобства восприятия
df['weekday'] = df['weekday'].replace({0 : 'Понедельник', 1 : 'Вторник', 2 : 'Среда',
                                       3 : 'Четверг', 4 : 'Пятница', 5 : 'Суббота', 6 : 'Воскресенье'})
```

### Этаж квартиры

Создадим функцию, которая присвоит категории этажам: первый, последний, другой.

```python
def floor_cat(row):
    if row['floor'] == 1:
        return 'Первый'
    elif row['floor'] == row['floors_total']:
        return 'Последний'
    return 'Другой'
    
df['floor_category'] = df.apply(floor_cat, axis=1)
```

### Сравнительный анализ квартир в центре Санкт-Петербурга

Проанализирем центр и изучим следующие параметры: площадь, цена, число комнат, высота потолков. Также выделим факторы, которые влияют на стоимость квартиры (число комнат, этаж, удалённость от центра, дата размещения объявления). Отличаются ли они от общих выводов по всему городу?

Создадим отфильтрованный датафрейм с расстоянием до центра в 8 километров.

```python
df_spb_center = df_spb.query('dist_to_center <= 8').reset_index(drop=True)
```

Сравним на гистограмме два датафрейма по Санкт-Петербургу и только по центру города.

###### Первый параметр для сравнения - общая площадь. Будем использовать сравнение гистограмм.
- `histtype = 'step'` - этот параметр удобен для сравнения двух гистограмм, области будут не закрашены и границы двух гистограмм будут четко видны
- `linewidth` - толщина линий для большей наглядности
- `density = True` - позволяет отобразить две гистограммы в одной системе координат и сделать наглядное сравнение
- `ax = ax` - позволяет отобразить две гистограммы на одной оси координат

```python
ax = df_spb.plot(kind = 'hist', y = 'total_area', histtype = 'step', bins = 20, range = (0, 200),
            alpha = 0.5, linewidth = 3, grid = True, density = True, figsize=(12,8), label = 'total_area' + ' spb')
df_spb_center.plot(kind = 'hist', y = 'total_area', histtype = 'step', ax = ax, linewidth = 3,
                   bins = 20, range = (0, 200), alpha = 0.8, grid = True, density = True, figsize=(12,8),
                   label = 'total_area' + ' center')
```

Как видим из сравнительной гистограммы средняя площадь квартир в центре составляет примерно 45-50 кв.м., также в центре намного меньше квартир малой площади до 40 кв.м. и больше квартир площадью от 70 кв.м.

Из этой гистограммы видно, что в центре цена выше, чем в целом по городу и среднее значение стоимости составляет около 7 млн.
